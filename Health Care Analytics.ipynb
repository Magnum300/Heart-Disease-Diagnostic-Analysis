# there are thirteen features in dataset

age:Person's age in years
sex: person's gender (1= male,0=female)
cp: the chest pain experiences by them(1= typical angina,value2: atypical angina,value3: non-anginal pain,value 4: asymptomatic )
trestbps: Person's resting blood pressure(mm hg on admission to the hospital)
chol: the person's cholestrol measurement in mg/dl
fbs: the person's fasting blood sugar level(>120 mg/dl,1=true;0=false)
restecg: Resting electrocardiographic measurement (0=normal,1=having ST-T wave abnormally,2=showing probable or definite left venticular hypertrophy by Estes' criteria)
thalach: The person's maximum heart rate achived
exang: Exercise induced angina(1=yes;0=no)
oldpeak: ST depression induced by exercise relative to rest
slope: the slope of the peak exercise ST segment (Value1: upsloping,Value2:flat,value3:Downsloping)
ca:The number of major vessels(0-3)
thal: A blood disorder called thalassemia(3=normal;6=fixed defect;7=reversable defect)
num:heart disease(0=no,1=yes)

# Percentage of people having heart disease

```python
num=data.groupby('num').size()
print(num)
```

```python
#Converting Numerical data into Categorical data
def heart_disease(row):
    if row ==0:
        return'Absence'
    elif row==1:
        return'Presence'
```

```python
#Applying converted data into our dataset eith new column-Heart_Disease

data['Heart_Disease']=data['num'].apply(heart_disease)
data.head()
```

```python
hd=data.groupby('Heart_Disease')['num'].count()
hd
```

```python
#Pie chart Creation of heart Disease Population % using matplotlib
plt.figure(figsize=(10,7))
plt.pie(hd,labels=['Absence','Presence'],autopct='%0.2f%%')
plt.title('Heart Disease Population %',fontsize=20)
plt.show()
```

As we can see from above pie chart- From the overall population,people having heart disease(45.87%)are lesser than those who don't have heart disease(54.13%)

```python
#Countplot of population age using matplotlib and seaborn

plt.figure(figsize=(15,7))
sns.countplot(x='age',data=data)
plt.title('Population Age',fontsize=17)
plt.xlabel('Age',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
data.describe()
```

As we can see from above, the minimum age is 29,max age is 77 and mean age is 54.43

```python
#Converting Numerical Data into Categorical Data
def gender(row):
    if row==1:
        return 'Male'
    elif row==0:
        return 'Female'
```

```python
#Applying converted data into our dataset with new column- sex1
data['sex1']=data['sex'].apply(gender)
data.head()
```

```python
#Converting Numerical data into categorical data

def age_range(row):
    if row>=29 and row<40:
        return 'Young Age'
    elif row>=40 and row<55:
        return'Middle Age'
    elif row>55:
        return 'Elder Age'
```

```python
#Applying converted data into out dataset with new column- Age_Range
data['Age_Range']=data['age'].apply(age_range)
data.head()
```

```python
#Swarm Plot Creation of Gender Based Age category using Matplotlib and Seaborn

plt.figure(figsize=(10,7))
sns.swarmplot(x='Age_Range',y='age',hue='sex1',data=data,order=['Young Age','Middle Age','Elder Age'],palette='Oranges_r',s=5)
plt.title('Gender Based Age Category',fontsize=17)
plt.xlabel('Age Category',fontsize=15)
plt.ylabel('Age',fontsize=15)
plt.show()
```

In Our Population Number of males are more in the middle age category and females are more in the Elder Age Category

```python
#Count Plot Creation of heart disease based on Age Category usiing Matplotlib and Seaborn
plt.figure(figsize=(7,5))
hue_order=['Young Age','Middle Age','Elder Age']
sns.countplot(x='Heart_Disease', hue='Age_Range',data=data,order=['Presence','Absence'],hue_order=hue_order,palette='Pastel1')
plt.title('Heart Disease Based On Age category',fontsize=17)
plt.xlabel('Heart Disease',fontsize=15)
plt.ylabel('Counts',fontsize=15)
plt.show()             
```

Elder age People are most affected by Heart Disease and middle age people are mostly free from any kind of Heart Disease.
